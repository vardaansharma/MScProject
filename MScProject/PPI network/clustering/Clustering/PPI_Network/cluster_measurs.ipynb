```python
import numpy as np
import pandas as pd
import glob
```

```python
cluster_files = glob.glob("*.csv")
print(cluster_files)

clusterparam = pd.read_csv('../../../parameterFiles/clusteringAlg.csv',header=None)
print(clusterparam)
```

```python
# a = clusterparam.iloc[clusterparam[1] == 'louvain_communities.csv']
n = clusterparam.loc[clusterparam[1] == 'louvain_communities.csv',2]
print(n.index[0])

# print(str(n[2]))
```

```python
cluster_files
```

```python
cluster_stuff = pd.DataFrame()#columns=['Clustering_algo','Cn','smallest','biggest','mean','median','mode','variance','MAD','skew','kurtosis'],index=['Clustering_algo'])
for file_name in cluster_files:
    file = pd.read_csv(file_name,skiprows=1,sep='\t',header=None)
    file_coms = file[1].values
#     print(file_name)
#     algo_name = clusterparam.loc[clusterparam[1] == 'spectral_communities.csv',2]
    algo_name = clusterparam.loc[clusterparam[1] == file_name,2]
#     n = algo_name
#     print(algo_name[algo_name.index[0]])
    algo_name = algo_name[algo_name.index[0]]
#     finding Cn
    Cn = len(list(set(file_coms)))
    
#     list with each community size
    com_sizes = pd.Series(file_coms).value_counts()
    
#     smallest and biggest
    smallest = com_sizes.min()
    largest = com_sizes.max()
    
#     mean median mode
    mean = com_sizes.mean()
    median = com_sizes.median()
    mode = com_sizes.mode().tolist()
    no_of_mode = []
    for i in mode:
        no_of_mode.append(com_sizes.tolist().count(i))
    variance = com_sizes.std()
    MAD = com_sizes.mad()
    skew = com_sizes.skew()
    kurtosis = com_sizes.kurtosis().tolist()
    print(kurtosis)
    
#     adding to the df
    new_row = pd.Series([algo_name,Cn,smallest,largest,mean,median,mode,variance,MAD,skew,kurtosis])
#     print(new_row)
#     cluster_stuff = cluster_stuff.append(new_row,ignore_index=True)
    cluster_stuff = cluster_stuff.append({
        'Clustering_algo': algo_name,
        'Cn': Cn,
        'smallest': smallest,
        'largest': largest,
        'mean': mean,
        'median': median,
        'mode': mode,
        'std': variance,
        'MAD': MAD,
        'skew':skew,
        'kurtosis': kurtosis,
        'mode_count': no_of_mode
    }, ignore_index=True)
    
    
```
