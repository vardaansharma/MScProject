```python
import pandas as pd
```

```python


# induvidual enrichments
Dis_en = 'RESULTS/PPI_Network/sgG5/permute_p_values_HDO.csv'
BP_en = 'RESULTS/PPI_Network/sgG5/permute_p_values_GOBP.csv'
# DE_en = 'OUT/permute_p_values_DE_sgG5.csv'
DE_en = 'OUT/permute_p_values_Diff_exp.csv'


# comparison enrichment 
Dis_BP_en = 'OUT/Comparisions/overlapCOMS_BPvsDE.csv'
Dis_DE_en = 'OUT/Comparisions/overlapCOMS_BPvsDisb.csv'
BP_DE_en  = 'OUT/Comparisions/overlapCOMS_DEvsDis.csv'
```

# BP

```python
# Annotationfile = 'Annotations/flatfile_human_gene2HDO.csv'
Annotationfile = 'Annotations/flatfile.go.BP.csv'
anfile = pd.read_csv(Annotationfile,sep='\t',skiprows=[0],header=None)
# anfile.head()
# a = anfile.loc[anfile[0] == 'GO:0051591']
# a[1][a.index.values[0]]
thefile = pd.read_csv(BP_en,sep='\t')
```

```python
save_df = pd.DataFrame()
for i in range(len(thefile)):
    community_no = i + 1
    s_c = thefile.iloc[community_no - 1,:]
    df = pd.DataFrame()
    a = [list(s_c.filter(like='p-value'))][0]
    # print(a)
    pvals = [a[0]]
    b = list(s_c.filter(regex='p-value(?:[^\(alt\)]+)|p-value '))
    # print(len(a))
    # print(len(b))
    pvals = pvals + b
    # print(len(pvals))
    df['pv'] = pvals
    b = list(s_c.filter(regex='p.adjusted(?:[^\(alt\)]+)'))
    b1 = [list(s_c.filter(like='p.adjusted'))[0]]
    # print(b1[0])
    # print(len(b))
    # print(len(df))
    colnames = thefile.columns.values
    GOterms = [item for item in colnames if 'GO' in item]
    df['padj'] = b1 + b #list(s_c.filter(like='p.adjusted'))
    df['95'] = list(s_c.filter(like='95%'))
    df['actual'] = list(s_c.filter(like='actual'))
    df['exptected'] = list(s_c.filter(like='expected'))
    df['GO'] = GOterms
    sorted_df = df.sort_values(by=['pv'])
    print('Community number:' + str(community_no))
    print((sorted_df.index[0] * 12) + 3)
    print(sorted_df.head(10))
    print('\n----------------\n')
    
    
#     checkig the enrichment levels and saving the enrichment terms in a csv
    enriched_GO_terms = []
    enriched_GO_ids = []
    p_vals = []
    enriched = 'NO'
    en_df = sorted_df[sorted_df['pv'] < 0.05]
    if(len(en_df)):
#         print(en_df.head())
        for index, row in en_df.iterrows():
            GO_term = anfile.loc[anfile[0] == row['GO']]
            
#             print(GO_term[1])
            enriched_GO_terms.append(GO_term[1][GO_term.index.values[0]])
            enriched_GO_ids.append(row['GO'])
            p_vals.append(row['pv'])
        enriched = 'YES'
    
    else:
        print('No significant enrichment')
        
    new_row ={
        'community': community_no,
        'Significant enrichment': enriched,
        'Enriched GO terms': enriched_GO_terms,
        'Enriched GO IDs': enriched_GO_ids,
        'p values': p_vals
    }
    save_df = save_df.append(new_row, ignore_index=True)

save_df.to_csv('enrichment_analysis/GO_BP_enrichment.csv',index=False, columns=['community','Significant enrichment','Enriched GO terms','Enriched GO IDs','p values'])
save_df.head()
```
